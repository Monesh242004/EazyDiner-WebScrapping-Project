```python
restaurants_name = []
prices = []
area = []
reatings = []
discounts = []

for i in range(1,151):
    url = f'https://www.eazydiner.com/bengaluru/restaurants?page={i}'
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
    }
    response = requests.get(url, headers = headers)
    html_code = response.text
    soup = BeautifulSoup(html_code,'html.parser')
    container = soup.find_all('div', attrs = {'class': 'relative pointer'})

    for c in container:
        # package Name
        restaurants = c.find('h3', attrs = {'class': 'margin-0-auto all_unset'})
        if restaurants is None:
            restaurants_name.append(np.nan)
        else:
            restaurants_name.append(restaurants.text)

        # duration
        pri = c.find('div', attrs = {'class': 'flex align-v-center gap_5 relative'})
        if pri is None:
            prices.append(np.nan)
        else:
            prices.append(pri.text)

        # Area 
        loc = c.find('div', attrs = {'class': 'flex align-v-center gap_5 padding relative'})
        if loc is None:
            area.append(np.nan)
        else:
            area.append(loc.text)

        # details 
        rating = c.find('div', attrs = {'class': 'flex align-v-start flex-between padding-t-15 padding-b-15 margin-t-12 relative full-width'})
        if rating is None:
            reatings.append(np.nan)
        else:
            reatings.append(rating.text)
            
        #package price
        discount = c.find('div', attrs = {'class': 'relative font-16 bold white nowrap ellipsis'})
        if discount is None:
            discounts.append(np.nan)
        else:
            discounts.append(discount.text)
```

```python
print("restaurants_name: ", len(restaurants_name))
print("prices: ", len(prices))
print("area: ", len(area))
print("reatings: ", len(reatings))
print("discounts: ", len(discounts))
```

```python
df = pd.DataFrame({
    "Restaurants_Name" : restaurants_name,
    "Cost_of_two" : prices,
    "Areas" : area,
    "Reatings" : reatings,
    "Discounts" : discounts
})
df
```

```python
df.to_csv('eazydiner_webscraping_dataset.csv',index = False)
```

```python
df[['Cuisine','cost_of_two']] = df['Cost_of_two'].str.extract(r'(.*?)(â‚ą.*)')
df['cost_of_two'] = df['cost_of_two'].str.extract(r'(\d+)').astype('int')
df
```

```python
df['ratings'] = df['Reatings'].str.extract(r'(\d\.\d)$').astype('float')
df
```

```python
df[['area', 'region']] = df['Areas'].str.extract(r"([^'\n]+?)(?:,\s*|\s+)((?:Central|West|South|North|East)?\s*Bengaluru)")
```

```python
df
```

```python
def extract_discounts(offer):
    return re.findall(r'(\d+)%', str(offer))

df['Discount'] = df['Discounts'].apply(extract_discounts)
df['Discount'].apply(pd.Series)
```

```python
df['Discount'] = df['Discount'].apply(lambda x: sum(map(int, x)) if isinstance(x, list) else 0)
df['Discount'].head()
```
